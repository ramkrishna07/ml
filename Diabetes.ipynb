```python
import pandas as pd
diabetes=pd.read_csv("https://github.com/YBI-Foundation/Dataset/raw/main/Diabetes.csv")
diabetes
```

```python
y=diabetes['diabetes']
X=diabetes.drop(['diabetes'],axis=1)
X
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.scatter(diabetes.bmi,y,marker='+',color='red')
```

```python
# splitting the dataset into 2 parts
# one for model training and  the other is for model testing
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.9,random_state=2529)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=500)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
# accuracy= (number of correct prediction)/(total number of prediction)
accuracy_score(y_test,y_pred)
```
